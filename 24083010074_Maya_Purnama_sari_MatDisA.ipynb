#LANGKAH 1 : Mengupload data dan memberikan data nilai random

pada langkah satu ini, code ini digunkana untuk mengupload file csv sebagai data awal, lalu data akan dibaca dan dispam dalam dataframe, saya menambahkan code untuk membenarkan index data, mengecek nama kolom asli , menambhakan code untuk mneghapus kolom yang kosong/typo , dan merename agar tamilan dataframe lebih rapi. lalu saya memberikan code untuk mengethuai kolom mana yang akam diisi oleh angka random yang dimulai dari angka 0.0-1.0. lalu diakhiri dengan menampilkan hasil dataframe


```python
import numpy as np
import pandas as pd
from google.colab import files

uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])


df.index = range(1, len(df) + 1)

print("Kolom asli:", df.columns.tolist())

df = df.drop(columns=[col for col in df.columns if "krediabel" in col.lower()])

df = df.rename(columns={
    "NPM": "NPM",
    "NAMA": "NAMA",
    "adil, kredibel (A)": "Adil, Kredibel (A)",
    "suka membantu (B)": "Suka Membantu (B)",
    "nyambung ketika ngobrol (C)": "Nyambung ketika ngobrol (C)",
    "asyik, menarik (D)": "Asyik, Menarik (D)"
})

kolom_nilai = ["Adil, Kredibel (A)", "Suka Membantu (B)",
               "Nyambung ketika ngobrol (C)", "Asyik, Menarik (D)"]

np.random.seed(45)
for kolom in kolom_nilai:
    df[kolom] = np.random.choice(np.arange(0, 1.1, 0.1), size=len(df))

display(df)
```

#Langkah 4 : Matriks Kovarian dan Matriks Korelasi
Pada langkah 4  ini, akan dihitung nilai matriks kovarian dan matriks korelasi.
Matriks kovarian digunakan untuk memahami hubungan antar variabel dalam satu set data, seperti menunjukkan seberapa kuat dua variabel berubah bersamaan (variabel yang meningkat bersama atau menuurn secara keterbalikan). Sedangkan Matriks korelasi digunakan untuk menganalisis dan memvisualisasikan hubungan linear antar variabel dalam sebuah kumpulan data.


```python
kelasmatdisA = df[kolom_nilai].to_numpy(dtype=float)

cov_mat = np.cov(kelasmatdisA, rowvar=False, ddof=1)
corr_mat = np.corrcoef(kelasmatdisA, rowvar=False)

print("Matriks Kovarian:\n", cov_mat, "\n")
print("Matriks Korelasi:\n", corr_mat)
```

Hasil Matriks Kovarian di atas menunjukkan bahwa Matriks kovarian berukuran 4×4 karena menghitung hubungan antara empat variabel, yaitu A, B, C, dan D. Nilai pada diagonal (0.0984, 0.1082, 0.0837, dan 0.0903) menunjukkan varian dari masing-masing variabel, yang berarti keempat variabel memiliki sebaran data yang hampir sama. Nilai di luar diagonal menunjukkan kovarian antar variabel. Kovarian antara A dan B sebesar -0.0026 menandakan hubungan sangat lemah dan berketerbalikan arah. Kovarian antara A dan C sebesar -0.0291 juga menunjukkan hubungan lemah dan berlawanan arah, sementara antara A dan D sebesar -0.0152 hampir tidak menunjukkan hubungan.

Secara keseluruhan nilai tersebut dapat disimpulkan bahwa hubungan antar variabel A, B, C, dan D relatif sangat lemah, baik yang searah maupun berlawanan arah.

#Langkah 5 : Cosine Similarity
Pada langkah 5  ini, akan dilakukan perhitungan Cosine similarity. Cosine similarity adalah metode untuk mengukur tingkat kemiripan antara dua objek berdasarkan sudut kosinus dari vektor data keduanya. Nilainya berada pada rentang -1 hingga 1, di mana nilai 1 menunjukkan dua objek identik, nilai mendekati 1 berarti sangat mirip, nilai mendekati 0 berarti hampir tidak ada kemiripan, dan nilai negatif berarti berlawanan arah.

```python
from numpy.linalg import norm

X_matrix = df[kolom_nilai].to_numpy(dtype=float)

students = df["NAMA"]

norms = np.linalg.norm(X_matrix, axis=1, keepdims=True)
cosine_sim = (X_matrix @ X_matrix.T) / (norms @ norms.T)

cosine_df = pd.DataFrame(cosine_sim, index=students, columns=students)

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

display(cosine_df)
```

Hasil perhitungan ditampilkan dalam bentuk matriks, di mana setiap baris dan kolom mewakili seorang mahasiswa. Misalnya, cosine similarity antara Aquina dan Raveena  sebesar 0.88 menunjukkan bahwa pola nilai keduanya mirip, atau dengan sekitar 88% kesamaan. Informasi ini berguna untuk melihat siswa yang memiliki kecenderungan nilai yang serupa pada variabel A, B, C, dan D, sehingga dapat diidentifikasi kelompok siswa dengan karakteristik penilaian yang hampir sama.

```python
#ANALISIS TAMBAHAN ( PCA dan Clustering)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df[kolom_nilai].to_numpy(dtype=float)
names = df["NAMA"].astype(str).values

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

Kode di atas digunakan untuk menyiapkan data sebelum dilakukannya analisis PCA dan Clustering. Pertama, nilai dari kolom yang dipilih (kolom_nilai) diambil lalu diubah menjadi matriks numerik. Nama mahasiswa disimpan dalam variabel name agar bisa digunakan kembali pada hasil analisis. Selanjutnya, data distandarisasi menggunakan StandardScaler, yaitu mengubah setiap variabel agar memiliki rata-rata nol dan standar deviasi satu. Proses standarisasi ini penting supaya semua variabel berada pada skala yang sama, sehingga analisis PCA dan K-Means tidak bias terhadap variabel yang memiliki rentang nilai lebih besar.

PCA (Principal Component Analysis) adalah teknik statistik yang digunakan untuk mereduksi dimensi data dengan cara mengubah banyak variabel yang saling berkorelasi menjadi sejumlah kecil variabel baru.pada code ini digunakan untuk menyederhanakan 4 variable yang tersedia menjadi 2 komponen agar mudah di visualisasi.

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_std)

print("Rasio varians per komponen:", pca.explained_variance_ratio_)
print("Total varians (PC1+PC2):", pca.explained_variance_ratio_.sum())

loadings = pd.DataFrame(
    pca.components_.T,
    index=kolom_nilai,
    columns=["Loading_PC1", "Loading_PC2"]
)
print("Loading (kontribusi tiap variabel ke komponen):")
display(loadings)

pca_df = pd.DataFrame({
    "NAMA": names,
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1]
})
display(pca_df.head())

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], s=35, alpha=0.8)
plt.title("PCA (2 Komponen) dari Variabel A–D")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

Hasil PCA menunjukkan bahwa dua komponen utama (PC1 dan PC2) mampu menjelaskan 67,26% variasi data, dengan PC1 menyumbang 39.99% dan PC2 sebesar 27,27%. Artinya, sebagian besar pola dari empat variabel (A: adil, kredibel; B: suka membantu; C: nyambung ketika ngobrol; D: asyik, menarik) sudah dapat dirangkum hanya dengan dua komponen ini.

Berdasarkan nilai loading, PC1 terutama dipengaruhi positif oleh A (0,45), dan D (0,55), serta sedikit negatif oleh B (-0,46) Dengan demikian, mahasiswa dengan skor PC1 tinggi cenderung menonjol pada aspek A dan D . Sementara itu, PC2 terutama dipengaruhi oleh A (0,592) dan sebagian oleh B (0,592), sehingga skor PC2 tinggi menggambarkan mahasiswa yang menonjol pada aspek A dan B.

Scatter plot memperlihatkan distribusi mahasiswa dalam bidang dua dimensi berdasarkan PC1 dan PC2. Titik-titik yang berdekatan menunjukkan mahasiswa dengan pola nilai mirip, sedangkan titik yang jauh terpisah mencerminkan perbedaan karakteristik yang besar antar mahasiswa.

Clustering ( K_Means dengan k=3)
 metode ini digunakan untuk mengelompokan data ke dalam 3 cluster berdasar kemiripan nilai pada masing-masing variabel. hasil dari metode ini yakni pembagian kelompok , rata-rata nilai cluster, serta visualisasi sebaran data dalam ruang pca

```python
k = 3
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_std)

df_clusters = df.copy()
df_clusters["Cluster"] = labels
print("Jumlah anggota per cluster:")
display(df_clusters["Cluster"].value_counts().sort_index())

centroids_std = kmeans.cluster_centers_
centroids_original = scaler.inverse_transform(centroids_std)
centroids_df = pd.DataFrame(centroids_original, columns=kolom_nilai)
centroids_df.index = [f"Cluster {i}" for i in range(k)]
print("Centroid tiap cluster (skala asli A–D):")
display(centroids_df)

plt.figure(figsize=(7,5))
for c in range(k):
    idx = labels == c
    plt.scatter(X_pca[idx,0], X_pca[idx,1], s=50, alpha=0.9, label=f"Cluster {c}")
plt.title("K-Means (k=3) divisualisasikan pada ruang PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

Hasil clustering dengan K-Means membagi mahasiswa menjadi tiga kelompok. Cluster 0 berjumlah 15 orang dengan karakteristik nilai rata-rata rendah pada aspek suka membantu (0.32), namun tinggi pada asik dan menarik (0,78),adil dan kredibel (0.62) dan nyambung kalau ngobrol (0,62). Cluster 1 terdiri dari 18 orang dengan nilai rata-rata cukup tinggi pada nyambung ketika ngobrol (0.66), tetapi rendah pada asik dan menarik (0,27), suka membantu(0,60), dan adil dan kredibel (0,36). Sementara itu, Cluster 2 berjumlah 7 orang dengan rata-rata nilai tinggi hampir di semua aspek, yaitu adil, kredibel (0,91), suka membantu (0,90), nyambung kalau ngobrol (0,30), dan asyik, menarik (0,34). Visualisasi pada ruang PCA menunjukkan pemisahan antar cluster cukup jelas, di mana setiap warna mewakili kelompok dengan pola karakteristik yang mirip.
